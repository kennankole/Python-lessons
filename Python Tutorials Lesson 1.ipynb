## Basic Data Types in Python

#### 1. Text Type:		str (string)
    
        name = "anko"
        name = 'anko'

#### 2. Numeric Types:	

__a__: int

      age = 9
      
__b__: float
     
     price = 10.50
     
__c__: complex

      r = 3+5j

#### 3. Sequence Types:	

__a__: list

       items = ['oranges', 'mangoes', 45, 10.50]

__b__: tuple

        items = ('oranges', 'mangoes', 45, 10.50)
        
__c__: range 

        items = range(40)

#### 4. Mapping Type: dict

        items = {'oranges':10, 'mangoes':15, 'name':'anko'}

#### 5. Set Types:	
__a__: set

        items = {3, 5, 6, 7, 9}
        #this set is mutable 
        items.add(500) #this is possible
        

__b__: frozenset
       is an immutable set
       
       new_items = frozenset(items) # new_items is now immutable
       new_items.add(500)#this is not possible

#### 6. Boolean Type:	bool

        are_you_sick = False
        are_you_healthy = True
